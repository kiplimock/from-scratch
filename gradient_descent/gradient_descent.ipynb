### Gradient Descent Algorithm

An optimisation algorithm for computing the local minima of a cost function. It is commonly used in machine learning to learn the optimal weights of a model.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

### 1. 2D visualisation using a polynomial

```python
def y_function(x):
    return x ** 2

def y_derivative(x):
    return 2 * x
```

```python
x = np.arange(-100, 100, 0.1)
y = y_function(x)

current_pos = (80, y_function(80))
learning_rate = 0.1

for _ in range(100):
    # clear the current output
    clear_output(wait=True)

    # create a new figure
    plt.figure()
    plt.plot(x, y, label="$y = x^2$")
    plt.scatter(current_pos[0], current_pos[1], color='r', label="Current Position")
    
    plt.legend()
    plt.title("Gradient Descent Visualisation")
    
    # display the plot
    plt.show()

    new_x = current_pos[0] - learning_rate * y_derivative(current_pos[0])
    new_y = y_function(new_x)
    
    current_pos = (new_x, new_y)
```

### 2. 2D visualisation using a sinusoid

```python
def y_function(x):
    return np.sin(x)

def y_derivative(x):
    return np.cos(x)
```

```python
x = np.arange(-5, 5, 0.1)
y = y_function(x)

current_pos = (1.5, y_function(1.5))
learning_rate = 0.1

for _ in range(100):
    # clear the current output
    clear_output(wait=True)

    # create a new figure
    plt.figure()
    plt.plot(x, y, label="$y = \sin{x}$")
    plt.scatter(current_pos[0], current_pos[1], color='r', label="Current Position")
    
    plt.legend()
    plt.title("Gradient Descent Visualisation")
    
    # display the plot
    plt.show()

    new_x = current_pos[0] - learning_rate * y_derivative(current_pos[0])
    new_y = y_function(new_x)
    
    current_pos = (new_x, new_y)
```

### 3D visualisation

```python
def z_function(x, y):
    return np.sin(5 * x) * np.cos(5 * y) / 5

def calculate_gradient(x, y):
    return np.cos(5 * x) * np.cos(5 * y), -np.sin(5 * y) * np.sin(5 * y)
```

```python
%matplotlib inline
```

```python
x = np.arange(-1, 1, 0.05)
y = np.arange(-1, 1, 0.05)

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

current_pos = (0.7, 0.4, z_function(0.7, 0.4))
learning_rate = 0.01



for _ in range(100):
    # clear the output
    clear_output(wait=True)

    # display the plot
    ax = plt.subplot(projection='3d', computed_zorder=False)
    ax.plot_surface(X, Y, Z, cmap='viridis', zorder=0)
    ax.scatter(current_pos[0], current_pos[1], current_pos[2], color='magenta', zorder=1)
    plt.show()

    # update X and Y
    X_derivative, Y_derivative = calculate_gradient(current_pos[0], current_pos[1])
    new_X = current_pos[0] - learning_rate * X_derivative
    new_Y = current_pos[1] - learning_rate * Y_derivative
    current_pos = (new_X, new_Y, z_function(new_X, new_Y))
```
